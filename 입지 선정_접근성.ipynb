```python
import pandas as pd
import numpy as np

df = pd.read_excel('C:/데이터/2018년 교통접근성지표_20210804/01_평균접근시간_2018.xlsx')

df.head()
```

```python
df = df.drop([0,1,2])
df = df.rename(columns=df.iloc[0])
df= df.drop([3])

df.head()
```

```python
df1 = df[df.HDCD_SGG_NM != '-']
df1 = df1.drop(['HDCD_EMD_NM','Region_NM'], axis=1)

df1.head()
```

```python
df1 = df1[df1.HDCD_SGG_NM != '-']
df2 = df1[df1['Faci_CA'] == '의료시설']

df2.head()
```

```python
df2.dtypes
```

```python
df2['HDCD_SGG_NM'].unique()
```

```python
df2 = df2.replace('>120.00','120.00')
```

```python
df3 = df2.astype({'평균접근시간(분)':'float'})

df3.head()
```

```python
df3.dtypes
```

```python
df4 = df3.pivot_table(index = ['HDCD_SGG_NM','Faci_NM','Mode_NM'], values = '평균접근시간(분)', aggfunc='mean')
df4 = df4.sort_values(by='평균접근시간(분)', ascending = False)

df4.head(15)
```

```python
df4.to_csv('C:/2018년 교통접근성지표_20210804/의료기관 접근성.csv', encoding = 'euc-kr')
```
