```python
import pandas as pd
import numpy as np

df = pd.read_csv('C:/데이터/시군구 성별 연령.csv', encoding = 'euc-kr')

df
```

```python
df1 = df.dropna(axis=1)

df1
```

```python
df2 = df1.replace('0 - 4세','01')
df2 = df2.replace('5 - 9세','02')
df2 = df2.replace('10 - 14세','03')
df2 = df2.replace('15 - 19세','04')
df2 = df2.replace('20 - 24세','05')
df2 = df2.replace('25 - 29세','06')
df2 = df2.replace('30 - 34세','07')
df2 = df2.replace('35 - 39세','08')
df2 = df2.replace('40 - 44세','09')
df2 = df2.replace('45 - 49세','10')
df2 = df2.replace('50 - 54세','11')
df2 = df2.replace('55 - 59세','12')
df2 = df2.replace('60 - 64세','13')
df2 = df2.replace('65 - 69세','14')
df2 = df2.replace('70 - 74세','15')
df2 = df2.replace('75 - 79세','16')
df2 = df2.replace('80 - 84세','17')
df2 = df2.replace('85 - 89세','18')

df2
```

```python
df3 = df2[(df2.연령별 != '계') & (df2.연령별 != '80세 이상') & (df2.연령별 != '85세 이상') & (df2.연령별 != '90 - 94세') & (df2.연령별 != '95 - 99세') & (df2.연령별 != '100세 이상')]

df3
```

```python
df4 = df3[df3['행정구역(시군구)별'] != '전국']
                 
df4
```

```python
df5 = df4.pivot_table(index = ['행정구역(시군구)별','성별','연령별'], values = '2020 년', aggfunc='sum')
df6 = df5.sort_values(by='2020 년', ascending = False)

df6
```
