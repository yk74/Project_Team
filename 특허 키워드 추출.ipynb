```python
import pandas as pd

file = open("./database_eated2.txt", mode='r', encoding='CP949')
data = file.read()

data
```

```python
from konlpy.tag import Kkma

kkma = Kkma()

ex_sent = kkma.sentences(data)
ex_nouns = kkma.nouns(data)
ex_nouns
```

```python
from re import match

nouns = []
for sent in ex_sent : 
    for noun in kkma.nouns(sent) :
        # 단어 전처리 : 2음절 이상, 수사 제외
        if len(str(noun)) >= 2 and not(match('^[0-9]', noun)) :
            nouns.append(noun)

nouns
```

```python
from re import match

ex_sent = kkma.sentences(data)
ex_pos = kkma.pos(data) 

nouns = []
for sent in ex_sent : 
    for noun in kkma.nouns(sent) : 
        if len(str(noun)) >= 2 and not(match('^[0-9]', noun)) :
            nouns.append(noun)
             
text_data = [] 
for (text, tclass) in ex_pos : 
    if tclass == 'NNG' or tclass == 'NNP' : 
        text_data.append(text)
        
ins0 = list(set(nouns).intersection(text_data))
ins0     
```

```python
ex_pos = kkma.pos(data) 

text_data = [] 
for (text, tclass) in ex_pos : # ('형태소', 'NNG') 
    if tclass == 'NNG' or tclass == 'NNP' : 
        text_data.append(text)

text_data
```

```python
ins = list(set(nouns).intersection(text_data))
ins
```

```python
with open("특허 키워드.txt", "w") as output:
    for row in ins:
        output.write(str(row) + '\n')
```
