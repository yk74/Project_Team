```python
import pandas as pd
import numpy as np

df = pd.read_csv('C:/데이터/T20_2019.csv', encoding = 'cp949')

df
```

```python
dt = pd.read_csv('C:/데이터/시도별 연령별 인구수.csv', encoding = 'cp949')

dt.head()
```

```python
df1 = df[(df['서식코드'] == 3)] # 의과 외래
```

```python
서울 = dt[dt['시도코드'] == 11]  # 서울
서울 = 서울.sort_values(by='인구수', ascending = False)

서울.head()
```

```python
서울1 = df1[(df1['시도코드'] == 11) & ((df1.연령대코드 == 6) | (df1.연령대코드 == 10) | (df1.연령대코드 == 11))] 
a = 서울1['진료과목코드'].value_counts()
서울1 = pd.DataFrame(a)
서울2 = 서울1.head()
서울2.rename(columns={'진료과목코드':'인구수'}, inplace = True)
서울2.insert(0, '시도코드', '11')
서울2['진료과목코드'] = 서울2.index
서울2.reset_index()

서울2.to_csv('C:/데이터/진료과목 선정.csv', encoding = 'euc-kr', index = None)
```
