```python
import pandas as pd
import numpy as np

df = pd.read_csv('C:/데이터/시군구_성_연령.csv', encoding = 'euc-kr')

df
```

```python
df.columns = ['시점','시도코드','성별코드','계','01','02','03','04','05','06','07','08','09','10','11','12','13','14','15','16','80세 이상','17','18','x1','x2','x3','x4']

df
```

```python
df1 = df.drop(['계','시점','80세 이상','x1','x2','x3','x4'], axis=1)


df1
```

```python
df2 = df1.replace('서울특별시','11')
df2 = df2.replace('부산광역시','26')
df2 = df2.replace('대구광역시','27')
df2 = df2.replace('인천광역시','28')
df2 = df2.replace('광주광역시','29')
df2 = df2.replace('대전광역시','30')
df2 = df2.replace('울산광역시','31')
df2 = df2.replace('세종특별자치시','36')
df2 = df2.replace('경기도','41')
df2 = df2.replace('강원도','42')
df2 = df2.replace('충청북도','43')
df2 = df2.replace('충청남도','44')
df2 = df2.replace('전라북도','45')
df2 = df2.replace('전라남도','46')
df2 = df2.replace('경상북도','47')
df2 = df2.replace('경상남도','48')
df2 = df2.replace('제주특별자치도','49')
df2 = df2.replace('남자','1')
df2 = df2.replace('여자','2')

df2.head()
```

```python
df3 = df2[df2['시도코드'] != '전국']
df3 = df3[df3['성별코드'] != '계']
                 
df3.head()
```

```python
df4 = df3.melt(id_vars=["시도코드", "성별코드"], 
        var_name="연령대코드", 
        value_name="인구수")

df4.head()
```

```python
df4.to_csv('C:/데이터/시도별 연령별 인구수.csv', encoding = 'euc-kr', index = None)
```
